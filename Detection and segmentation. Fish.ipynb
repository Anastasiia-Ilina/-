```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import vgg16
from glob import glob
import zipfile
import os
from random import shuffle
import re

from matplotlib import pyplot as plt
```

# Загрузим данные

```python
train = zipfile.ZipFile('train.zip', 'r')
test = zipfile.ZipFile('test.zip', 'r')
```

```python
train.extractall()
test.extractall()
```

```python
train
```

```python
train_files = glob('train/*.jpg')
test_files = glob('test/*.jpg')
```

```python
# проведем педобработку для модели
def load_image(path, target_size=(224, 224)):
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, target_size)
    return vgg16.preprocess_input(img)
```

```python
def fit_generator(files, batch_size=32):
    batch_size = min(batch_size, len(files))
    while True:
        shuffle(files)
        for k in range(len(files) // batch_size):
            i = k * batch_size
            j = i + batch_size
            if j > len(files):
                j = len(files)
            x = np.array([load_image(path) for path in files[i:j]])
            y = np.array([1. if os.path.basename(path).startswith('dog') else 0.
                          for path in files[i:j]])
            yield (x, y)
```

```python
def predict_generator(files):
    while True:
        for path in files:
            yield np.array([load_image(path)]),
```

# Посмотрим на изображения

```python
fig = plt.figure(figsize=(16, 8))
for i, path in enumerate(train_files[:10], 1):
    subplot = fig.add_subplot(2, 5, i)
    subplot.set_title('%s' % path.split('/')[-1])
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, (224, 224))
    plt.imshow(img)
```

# Загрузим уже обученную модель

```python
base_model = vgg16.VGG16(weights='imagenet',
                         include_top=False,
                         input_shape=(224, 224, 3))

base_model.summary()
```

# Добавим свой слой

```python
for layer in base_model.layers:
    layer.trainable = False

x = base_model.layers[-5].output
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
x = tf.keras.layers.Dense(1, activation='sigmoid',
                          kernel_regularizer=tf.keras.regularizers.l1(1e-4)
                          )(x)

model = tf.keras.Model(inputs=base_model.input, outputs=x, name='dogs_vs_cats')
```

```python
model.summary()
```

# Перейдем к обучению

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
val_samples = 5
shuffle(train_files)
validation_data = next(fit_generator(train_files[:val_samples], val_samples))
train_data = fit_generator(train_files[val_samples:])


model.fit(train_data,
          steps_per_epoch=10,
          epochs=20,
          validation_data=validation_data)
```

```python
model.save('cats-dogs-vgg16.keras')
```

```python
test_pred = model.predict(predict_generator(test_files), steps=len(test_files))
```

```python
fig = plt.figure(figsize=(16, 8))
for i, (path, score) in enumerate(zip(test_files[:10], test_pred[:10]), 1):
    subplot = fig.add_subplot(2, 5, i)
    subplot.set_title('%.2f %s' % (score[0], os.path.basename(path)))
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, (224, 224))
    subplot.imshow(img)
```

# Сохраним данные

```python
with open('submit.csv', 'w') as dst:
    dst.write('id,label\n')
    for path, score in zip(test_files, test_pred):
        dst.write('%s,%f\n' % (re.search('(\d+).jpg$', path).group(1), score))
```
